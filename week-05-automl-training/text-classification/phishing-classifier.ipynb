```python
!pip install -q scikit-learn pandas
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
data = {
    'text': [
        # Phishing examples
        "Your account has been suspended. Click here immediately to verify your identity.",
        "We detected unusual activity on your bank account. Log in now to secure your funds.",
        "Your password expires today. Confirm your credentials to avoid losing access.",
        "You have won a gift card. Claim your prize by entering your login details.",
        "Urgent: your mailbox storage is full. Sign in now to keep receiving messages.",
        "We noticed a suspicious login attempt. Verify your account immediately.",
        "Your package delivery failed. Update your shipping information using the link below.",
        "Action required: confirm your payroll information to avoid payment delays.",
        "Your Netflix account is on hold. Update your billing information now.",
        "Security alert: unauthorized sign-in detected. Reset your password here.",
        "Final warning: failure to verify your account will result in permanent closure.",
        "Click this secure link to unlock your account before midnight.",
        "You must confirm your identity within 24 hours or your account will be disabled.",
        "We were unable to process your recent payment. Submit your card details again.",
        "A refund is waiting for you. Sign in to claim it now.",
        "Your organization requires immediate password confirmation for email access.",
        "This is your last chance to avoid account deactivation. Verify now.",
        "Your tax refund is ready. Provide your banking details to receive payment.",
        "An important encrypted document has been shared with you. Open it here.",
        "Confirm your login details now to restore full access to your account.",

        # Legitimate examples
        "Hi team, our meeting is scheduled for tomorrow at 10 AM in the main conference room.",
        "Your order has been shipped and is expected to arrive on Friday.",
        "Thank you for your purchase. Your receipt is attached for your records.",
        "This is a reminder that your dentist appointment is next Tuesday at 2 PM.",
        "Please review the attached agenda before our project check-in tomorrow.",
        "Your weekly report has been submitted successfully.",
        "Welcome to the course. Please review the syllabus before the first class.",
        "Your package has been delivered at the front door.",
        "Thank you for registering for the webinar. We look forward to seeing you there.",
        "The IT department will perform scheduled maintenance this weekend.",
        "Your prescription is ready for pickup at the pharmacy.",
        "Here is the updated calendar invite for Friday’s meeting.",
        "We received your payment successfully. No further action is needed.",
        "Please find attached the notes from today’s discussion.",
        "Your account settings were updated successfully through the official portal.",
        "Lunch with the team has been moved to 1 PM.",
        "Your library book is due in three days.",
        "This month’s utility bill is now available in your customer portal.",
        "Thank you for contacting support. Your request has been received.",
        "The assignment deadline has been extended until Sunday evening."
    ],
    'label': [
        'phishing','phishing','phishing','phishing','phishing',
        'phishing','phishing','phishing','phishing','phishing',
        'phishing','phishing','phishing','phishing','phishing',
        'phishing','phishing','phishing','phishing','phishing',

        'legitimate','legitimate','legitimate','legitimate','legitimate',
        'legitimate','legitimate','legitimate','legitimate','legitimate',
        'legitimate','legitimate','legitimate','legitimate','legitimate',
        'legitimate','legitimate','legitimate','legitimate','legitimate'
    ]
}

df = pd.DataFrame(data)
print(f"Dataset: {len(df)} records")
print(df['label'].value_counts())
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df['text'],
    df['label'],
    test_size=0.2,
    random_state=42,
    stratify=df['label']
)

vectorizer = TfidfVectorizer(max_features=500)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

model = MultinomialNB()
model.fit(X_train_vec, y_train)

predictions = model.predict(X_test_vec)

print("Confusion Matrix:")
print(confusion_matrix(y_test, predictions, labels=['phishing', 'legitimate']))

print("\nClassification Report:")
print(classification_report(y_test, predictions, labels=['phishing', 'legitimate']))

print("Accuracy:", accuracy_score(y_test, predictions))
```

```python
new_texts = [
    "Your account will be locked unless you verify your password immediately.",
    "Thank you for attending today's meeting. Attached are the follow-up notes.",
    "We detected suspicious activity. Confirm your login credentials now.",
    "Your order confirmation and receipt are available in your account portal.",
    "Update your banking details today to avoid interruption of service."
]

new_vec = vectorizer.transform(new_texts)
results = model.predict(new_vec)
probs = model.predict_proba(new_vec)

for text, label, prob in zip(new_texts, results, probs):
    confidence = max(prob)
    print(f"Input: {text}")
    print(f"  Prediction: {label} (confidence: {confidence:.2f})\n")
```
